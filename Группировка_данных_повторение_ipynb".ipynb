```
#Функции unique и nunique 

#список уникальных элементов из серии unique()
a = df['Nationality'].unique()
display(a)
len(a)
```

```
#количество уникальных значений в серии nunique()
df['Nationality'].nunique()
```

```
#Преобразование серии value_counts в датафрейм
s6 = df['Nationality'].value_counts()
s_df = s6.reset_index()
display(s_df)
```

```
#переименуем столбцы получившегося датафрейма в соответствии с тем, что в них хранится:
s_df.columns = ['Nationality','Players Count']
display(s_df)
```
